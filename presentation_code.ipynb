# EDA Presentation Code

First, we will load in the dataset and make the necessary imports.

```python
# Necessary imports
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Data loading
data = pd.read_csv('iowa_sales.csv')
```

```python
# Examine columns and shape
datatypes = data.dtypes
print(datatypes)
print(data.shape[0])
print(data.shape[1])
```

Here we see the data is about liquor purchases in the state of Iowa for a certain time period. Each row represents the invoice for purchases of the same bottle (i.e. a row could be one bottle of vodka, or two bottles of the same tequila, etc.). Each column is a different attribute of the purchase, such as the date purchased, the store the purchase was made at, the number of bottles purchased, the cost of the purchase and the alcohol itself.

Now let's rename the column names to be more readable!

```python
# Rename columns for readability
data.columns = data.columns.str.lower().str.replace(' ', '_')
```

Now that we've renamed the column names to be in a more readable format, we can begin our analysis.

```python
# Convert the 'Date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Convert "Sale (Dollars)" column to numeric format
data['sale_(dollars)'] = data['sale_(dollars)'].astype(str).str.replace('$', '').astype(float)

# Group by 'Date' and sum the sales for each date
sales_by_date = data.groupby('date')['sale_(dollars)'].sum()

# Plotting the sales over time
plt.figure(figsize=(16, 8))
sales_by_date.plot()
plt.title('Sales over Time', fontsize=30)
plt.ylabel('Total Sales (Dollars)', fontsize=24)
plt.xlabel('Date', fontsize=24)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter for Johnson county
johnson_data = data[data['county'] == 'Johnson']

# Group by "Item Description" and sum the "Bottles Sold" for each item
popular_items = johnson_data.groupby('item_description')['bottles_sold'].sum()

# Find the top 5 items by sorting popular items
top_5_items = popular_items.sort_values(ascending=False).head(5)

# Plotting the top 5 items in a bar chart
plt.figure(figsize=(14, 8))
top_5_items.plot(kind='bar', color='skyblue')
plt.title('Top 5 Most Popular Items in Johnson County', fontsize=30)
plt.ylabel('Total Bottles Sold', fontsize=24)
plt.xlabel('Item Description', fontsize=24)
plt.xticks(rotation=45, ha='right', fontsize=20)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Calculating price per liter
johnson_data['price_per_liter'] = johnson_data['sale_(dollars)'] / johnson_data['volume_sold_(liters)']

# Sorting to top 5 cheapest values
cheapest_alcohols = johnson_data.sort_values(by='price_per_liter').head(10)

# Displaying the relevant columns for these items
cheapest_alcohols_table = cheapest_alcohols[['date', 'item_description', 'volume_sold_(liters)', 'sale_(dollars)', 'price_per_liter']]
cheapest_alcohols_table
```
